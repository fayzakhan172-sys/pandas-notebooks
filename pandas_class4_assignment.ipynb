
# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
# Write your code here
import pandas as pd

data = {
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "City": ["Bhopal", "Delhi", "Bhopal", "Delhi"],
    "Marks": [80, 70, 90, 60]
}

df = pd.DataFrame(data)

result = df.groupby("City")["Marks"].mean()

print(result)
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
import pandas as pd

data = {
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Age": [22, 19, 21, 23],
    "Marks": [85, 75, 90, 65]
}

df = pd.DataFrame(data)

# Multiple conditions
result = df[(df["Marks"]>70)&(df["Age"]>20)]
print (result)
````
